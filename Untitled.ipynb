```python
import pandas as pd
```

```python
df = pd.read_csv('data/clinfo.csv')
dates = set()
```

```python
dates |= set(df['date'].map(lambda s : s[:4]))
print dates
```

```python
dates = map(lambda s : int(s), dates)
```

```python
df['year'] = df['date'].map(lambda s : int(s[:4]))
```

```python
df_year = {}
for year in dates :
    df_year[year] = df[df['year'] == year]
    print year
```

```python
df.hist(column='year', bins=len(dates))
```

```python
cumsum = 0
for year in df_year :
    cumsum += len(df_year[year])
    print year,len(df_year[year]), cumsum
```

```python
df_year[1840] = pd.concat([df_year[year] for year in df_year if year in range(1000, 1841)])
```

```python
for key in list(df_year.keys()) :
    if key in range(1000, 1840) :
        del df_year[key]
```

```python
for key in df_year :
    df_year[key].to_csv('data/year_wise/data_'+str(key)+'.csv')
```

```python
joins = {}
for key in df_year :
    joins[key] = df_year[key].join(df_year[key], on='cluster', how='inner', lsuffix='_A', rsuffix='_B')
    joins[key] = joins[key][joins[key].series_A != joins[key].series_B]
    print key
```
